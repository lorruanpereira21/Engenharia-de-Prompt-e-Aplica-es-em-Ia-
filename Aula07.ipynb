```
def calculadora_simples():
    # Inicia um bloco try-except para lidar com possíveis erros durante a execução
    try:
        # Solicita o primeiro número ao usuário como string
        num1_str = input("Digite o primeiro número: ")
        # Solicita o segundo número ao usuário como string
        num2_str = input("Digite o segundo número: ")

        # Converte as strings dos números para float, permitindo decimais
        num1 = float(num1_str)
        num2 = float(num2_str)

        # Solicita a operação desejada e converte para minúsculas para padronização
        operacao = input("Digite a operação (soma ou subtracao): ").lower()

        # Verifica qual operação foi escolhida pelo usuário
        if operacao == 'soma':
            # Realiza a operação de soma
            resultado = num1 + num2
            # Imprime o resultado da soma
            print(f"Resultado da soma: {resultado}")
        elif operacao == 'subtracao':
            # Realiza a operação de subtração
            resultado = num1 - num2
            # Imprime o resultado da subtração
            print(f"Resultado da subtração: {resultado}")
        else:
            # Mensagem de erro para operação inválida
            print("Erro: Operação inválida. Por favor, use 'soma' ou 'subtracao'.")

    except ValueError:
        # Captura o erro se a conversão para float falhar (entrada não numérica)
        print("Erro: Entrada inválida. Por favor, digite apenas números.")
    except Exception as e:
        # Captura qualquer outro erro inesperado que possa ocorrer
        print(f"Ocorreu um erro inesperado: {e}")

# Chama a função calculadora_simples para iniciar a execução
calculadora_simples()
```
